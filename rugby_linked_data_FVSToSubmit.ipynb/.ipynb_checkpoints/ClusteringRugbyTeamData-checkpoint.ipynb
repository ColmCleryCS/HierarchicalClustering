```python
def addColumnsForLeague(football_teams_df):
    dataset_column_length = len(football_teams_df.columns)
    leaguesIndex = []
    leagues = ["superRugby","top14","urc","prem"]
    for i in range(len(leagues)):
        leaguesIndex.append((i+1,leagues[i]))
        
    for league in leagues:
        football_teams_df["League_"+ league]=football_teams_df["LeagueName"].map(lambda x: league == x)
    football_teams_df.iloc[:,dataset_column_length:] = football_teams_df.iloc[:,dataset_column_length:].astype(int)
    football_teams_df = football_teams_df.drop('LeagueName', 1)
    return football_teams_df

def getHeight(x):
    x = float(x)
    if x > 100:
        x = x/100
    
    if x > 1.5 and x < 2.2:
        return x
    else:
        return None
    
def validateBirthYear(x):
    x = float(x)
    if x > 1975:
        return x
    else:
        return None


def edit_data_frame_with_columns(lst_of_features,rugby_teams_df):
    if 'teamname' not in lst_of_features:
        rugby_teams_df = rugby_teams_df.drop('teamname', 1)
    if 'longitude' in lst_of_features:
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['longitude'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('longitude', 1)
    if 'latitude' in lst_of_features:
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['latitude'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('latitude', 1)
    if 'propsheight' in lst_of_features:  
        rugby_teams_df["propsheight"] = rugby_teams_df["propsheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['propsheight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('propsheight', 1)
    if 'propsweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['propsweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('propsweight', 1)
    if 'propsbirthDate'  in lst_of_features:  
        rugby_teams_df["propsbirthDate"] = rugby_teams_df["propsbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['propsbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('propsbirthDate', 1)
    if 'hookersheight' in lst_of_features:  
        rugby_teams_df["hookersheight"] = rugby_teams_df["hookersheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['hookersheight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('hookersheight', 1)
    if 'hookersweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['hookersweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('hookersweight', 1)
    if 'hookersbirthDate'  in lst_of_features:  
        rugby_teams_df["hookersbirthDate"] = rugby_teams_df["hookersbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['hookersbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('hookersbirthDate', 1)
    if 'locksheight' in lst_of_features:  
        rugby_teams_df["locksheight"] = rugby_teams_df["locksheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['locksheight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('locksheight', 1)
    if 'locksweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['locksweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('locksweight', 1)
    if 'locksbirthDate'  in lst_of_features:  
        rugby_teams_df["locksbirthDate"] = rugby_teams_df["locksbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['locksbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('locksbirthDate', 1)
    if 'flankersheight' in lst_of_features:  
        rugby_teams_df["flankersheight"] = rugby_teams_df["flankersheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['flankersheight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('flankersheight', 1)
    if 'flankersweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['flankersweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('flankersweight', 1)
    if 'flankersbirthDate'  in lst_of_features:  
        rugby_teams_df["flankersbirthDate"] = rugby_teams_df["flankersbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['flankersbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('flankersbirthDate', 1)
    if 'number_eightsheight' in lst_of_features:  
        rugby_teams_df["number_eightsheight"] = rugby_teams_df["number_eightsheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['number_eightsheight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('number_eightsheight', 1)
    if 'number_eightsweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['number_eightsweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('number_eightsweight', 1)
    if 'number_eightsbirthDate'  in lst_of_features:  
        rugby_teams_df["number_eightsbirthDate"] = rugby_teams_df["number_eightsbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['number_eightsbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('number_eightsbirthDate', 1)
    if 'scrum_halvesheight' in lst_of_features:  
        rugby_teams_df["scrum_halvesheight"] = rugby_teams_df["scrum_halvesheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['scrum_halvesheight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('scrum_halvesheight', 1)
    if 'scrum_halvesweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['scrum_halvesweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('scrum_halvesweight', 1)
    if 'scrum_halvesbirthDate'  in lst_of_features:  
        rugby_teams_df["scrum_halvesbirthDate"] = rugby_teams_df["scrum_halvesbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['scrum_halvesbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('scrum_halvesbirthDate', 1)
    if 'fly_halvesheight' in lst_of_features:  
        rugby_teams_df["fly_halvesheight"] = rugby_teams_df["fly_halvesheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['fly_halvesheight'], errors='coerce').notnull()]

    else:
        rugby_teams_df = rugby_teams_df.drop('fly_halvesheight', 1)
    if 'fly_halvesweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['fly_halvesweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('fly_halvesweight', 1)
        
    if 'fly_halvesbirthDate'  in lst_of_features:  
        rugby_teams_df["fly_halvesbirthDate"] = rugby_teams_df["fly_halvesbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['fly_halvesbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('fly_halvesbirthDate', 1)
         
    if 'centresheight' in lst_of_features:  
        rugby_teams_df["centresheight"] = rugby_teams_df["centresheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['centresheight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('centresheight', 1)
        
    if 'centresweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['centresweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('centresweight', 1)
        
    if 'centresbirthDate'  in lst_of_features:  
        rugby_teams_df["centresbirthDate"] = rugby_teams_df["centresbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['centresbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('centresbirthDate', 1)
        
    if 'wingersheight' in lst_of_features:  
        rugby_teams_df["wingersheight"] = rugby_teams_df["wingersheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['wingersheight'], errors='coerce').notnull()]

    else:
        rugby_teams_df = rugby_teams_df.drop('wingersheight', 1)
        
    if 'wingersweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['wingersweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('wingersweight', 1)
        
    if 'wingersbirthDate'  in lst_of_features:  
        rugby_teams_df["wingersbirthDate"] = rugby_teams_df["wingersbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['wingersbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('wingersbirthDate', 1)
         
    if 'full_backsheight' in lst_of_features:  
        rugby_teams_df["full_backsheight"] = rugby_teams_df["full_backsheight"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['full_backsheight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('full_backsheight', 1)
        
    if 'full_backsweight' in lst_of_features:  
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['full_backsweight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('full_backsweight', 1)
        
    if 'full_backsbirthDate'  in lst_of_features:  
        rugby_teams_df["full_backsbirthDate"] = rugby_teams_df["full_backsbirthDate"].apply(validateBirthYear)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['full_backsbirthDate'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('full_backsbirthDate', 1)
        
    if 'competitionsWon' in lst_of_features: 
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['competitionsWon'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('competitionsWon', 1)
     
    if 'LeagueName' in lst_of_features: 
        rugby_teams_df = addColumnsForLeague(rugby_teams_df) 
    else:
        rugby_teams_df = rugby_teams_df.drop('LeagueName', 1)
        
    if 'height' in lst_of_features: 
        rugby_teams_df["height"] = rugby_teams_df["height"].apply(getHeight)
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['height'], errors='coerce').notnull()]

    else:
        rugby_teams_df = rugby_teams_df.drop('height', 1)
    
    if 'weight' in lst_of_features: 
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['weight'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('weight', 1)
        
    if 'youngestBirthYear' in lst_of_features: 
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['youngestBirthYear'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('youngestBirthYear', 1)
        
    if 'oldestBirthYear' in lst_of_features: 
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['oldestBirthYear'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('oldestBirthYear', 1)

    if 'averageBirthAge' in lst_of_features: 
        rugby_teams_df = rugby_teams_df[pd.to_numeric(rugby_teams_df['averageBirthAge'], errors='coerce').notnull()]
    else:
        rugby_teams_df = rugby_teams_df.drop('averageBirthAge', 1)
    
    return rugby_teams_df


def find_largest_silhoutte_score_for_features(lst_of_features,football_teams_df,cosine_distance):
    football_teams_df = edit_data_frame_with_columns(lst_of_features,football_teams_df)
    scaler = MinMaxScaler()
    data_scaled = pd.DataFrame(scaler.fit_transform(football_teams_df), columns=football_teams_df.columns)
    data_scaled.head()
    distance_metric = 'euclidean'
    linkage_types = ['ward','average','complete','single']
    if cosine_distance == True:
        distance_metric = 'cosine'
        linkage_types = ['average','complete','single']
    max_clusters = 0
    if len(football_teams_df.index) > 17:
        max_clusters = 17
    else:
        max_clusters = int(len(football_teams_df.index)/2)
    silhoutte_scores = []
    for link in linkage_types:
        for i in range(2,max_clusters):
            cluster = AgglomerativeClustering(n_clusters=i, affinity=distance_metric, linkage=link) 
            model = cluster.fit_predict(data_scaled)
            hierarchical_clustering = pd.DataFrame(model)
            silhoutte_scores.append((link,i,silhouette_score(data_scaled,hierarchical_clustering[0],metric=distance_metric)))
            
    silhoutte_scores.sort(key=lambda x:x[2]) #Order elements by the Silhoutte Score
    return silhoutte_scores[-1]

def find_best_features(lst,num_features,football_teams_df,cosine_distance):
    largest_element = None
    players_df = football_teams_df.copy()
    i = 0
    current_elements = []
    optimal_num_cluster = 0
    optimal_link_type = ""
    while i < num_features:
        max_shil_score = 0
        largest_element = 0
        for items in lst:
            test_players_df = players_df.copy()
            new_lst = current_elements + [items]
            link, num_clusters ,score = find_largest_silhoutte_score_for_features(new_lst,test_players_df,cosine_distance)
            if score > max_shil_score:
                largest_element = items
                max_shil_score = score
                optimal_num_cluster = num_clusters
                optimal_link_type = link
        current_elements.append(largest_element)
        lst.remove(largest_element)
        i += 1
    return ((current_elements,optimal_link_type, optimal_num_cluster,max_shil_score))
```

```python
import time
num_clusters = 5
affin = 'euclidean'
link = 'ward'
rugby_teams_df = pd.read_excel("rugby_stats_for_clubs_and_teams.xlsx", sheet_name="Sheet1_Rugby_Teams")
df1_rugby_teams = edit_data_frame_with_columns(['latitude','longitude','teamname','finishingPosition', 'LeagueName', 'locksbirthDate', 'propsbirthDate'],rugby_teams_df)
df1_rugby_teams_copy = df1_rugby_teams.copy()
df1_rugby_teams.drop('longitude', inplace=True, axis=1)
df1_rugby_teams.drop('latitude', inplace=True, axis=1)
df1_rugby_teams.drop('teamname', inplace=True, axis=1)
scaler = MinMaxScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(df1_rugby_teams), columns=df1_rugby_teams.columns)
cluster = AgglomerativeClustering(n_clusters=num_clusters, affinity=affin, linkage=link) 
team_models = cluster.fit_predict(data_scaled)
hierarchical_clustering = pd.DataFrame(team_models)
data_scaled.insert((data_scaled.shape[1]),'agglomerative',hierarchical_clustering)
dct_of_colour =  {0:"red", 1:"green", 2:"blue", 3:"purple", 4:"orange"}
times_dct = {"Cluster ID":[],"Number Of Teams":[],"Time To Load":[],}
m = folium.Map(location=[51.5074, 0.1278], zoom_start=6)
cluster_id = 0
for cluster in range(num_clusters):
    i = 0
    t1 = time.time()
    count_per_cluster = 0
    for index, row in df1_rugby_teams_copy.iterrows():
        if team_models[i] == cluster_id:
            count_per_cluster += 1
            html="""
                    <h1>""" + str(row['teamname']) + """'s data</h1><br>
                    <p>
                    </p>
                    <b> Club Name: </b> """ + str(row['teamname']) + """ <br>
                    <b> Finishing Position: </b> """ +  str(row['finishingPosition']) + """ <br>
                    <b> Average Birth Year Of Locks at club: </b> """ + str(row['locksbirthDate']) + """ <br>
                    <b> Average Birth Year Of Props at club: </b> """ + str(row['propsbirthDate']) + """ <br>
                    """
            frame = IFrame(html=html, width=500, height=300)
            pop_text = folium.Popup(frame, max_width=2650)
            folium.Marker(
                    [row['latitude'], row['longitude']], 
                    popup = pop_text, 
                    icon=folium.Icon(color = dct_of_colour[hierarchical_clustering[0][i]], icon="fa-futbol"),
                    ).add_to(m)
        i += 1
    t2 = time.time()
    times_dct["Time To Load"].append(t2 - t1)
    cluster_id+=1
    times_dct["Number Of Teams"].append(count_per_cluster)
    times_dct["Cluster ID"].append(cluster_id)
df = pd.DataFrame(times_dct)
print(df.to_latex())
data = io.BytesIO()
m.save(data, close_file=False)
m



```
